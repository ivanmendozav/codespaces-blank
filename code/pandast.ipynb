```python
import pandas as pd
import numpy as np
```

Crear dataset de ejemplo

```python
frame = pd.DataFrame([[6, np.nan, 6],[np.nan, np.nan, np.nan],
                      [2, np.nan, 5]],index=['blue','green','red'],
                      columns = ['ball','mug','pen'])
frame
```

Elimnar filas con NaN

```python
frame.dropna(how='all', inplace=True)
frame
```

Estadisticas de datos agrupados

```python
frame2 = pd.DataFrame(np.random.randint(10,20,20).reshape(10,2),
                      columns=["col1","col2"], 
                      index=pd.date_range("2022-01-01", periods=10))
frame2["grupo"] = np.random.choice(["A","B"],10)
frame2
```

```python
frame2.groupby("grupo").sum()
```

```python
frame2.groupby("grupo").mean().sort_values("col1")
```

```python
frame2["grupo2"] = np.repeat(["C","D"],5)
frame2.groupby("grupo2").min().sort_values("col2")
```

Manipulacion

```python
frame2["id"] = np.arange(2,12,1,dtype=int)
frame2
```

Merge para juntar por columnas

```python
frame3 = pd.DataFrame(np.random.rand(30).reshape(10,3)*3, 
                      columns=["col11","col12","col13"])
frame3["id"] = np.arange(2,12,1,dtype=int)
frame3
```

```python
merged = pd.merge(frame2, frame3)[["id","col1","col11","col12","grupo2"]]
merged
```

```python
merged.columns = ["id","secuencia","peso","porte","grupo"]
merged
```

```python
frame2.columns = ["col1","col2","grupo","grupo2","colid"]
pd.merge(frame2,frame3,left_on="colid",right_on="id").iloc[:,5:]
```

Concatenar matrices

```python
array1 = np.arange(9).reshape((3,3))
array2 = np.arange(9).reshape((3,3))+6
np.concatenate([array1, array2], axis=1)
```

```python
df1 = pd.DataFrame(np.random.rand(9).reshape(3,3),  columns=['A', 'B', 'C'])
df2 = pd.DataFrame(np.random.rand(9).reshape(3,3),  columns=['A', 'B', 'C'])
df3 = pd.concat([df1, df2], axis=0)
df3.index = ["a1","a2","a3","a4","a5","a6"]
df3
```
